## Data Cleaning

### Replace missing 'Age' values with the median value

```python
data['Age'] = data['Age'].fillna(data['Age'].median())
```

### Replace missing 'Fare' values with the median value

```python
data['Fare'] = data['Fare'].fillna(data['Fare'].median())
```

### Fill missing 'Cabin' values with 'Unknown' to distinguish absent data


```python
data['Cabin'] = data['Cabin'].fillna('Unknown')
```

## Create a figure with two subplots for Age and Fare distributions


```python
plt.figure(figsize=(12, 6))
```

## Plot Age distribution using a histogram and density plot

```python
plt.subplot(1, 2, 1)
sns.histplot(data['Age'], bins=30, kde=True, color='blue')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
```

## Plot Fare distribution using a histogram and density plot

```python
plt.subplot(1, 2, 2)
sns.histplot(data['Fare'], bins=30, kde=True, color='green')
plt.title("Distribution of Ticket Fares")
plt.xlabel("Fare")
```

## Plot the graph with respect to gender

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Survived',hue='Sex',data=data)
plt.title('Survival by Gender ')
plt.show()
```

## Plot the graph with respect to Passenger class

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=data, palette='plasma')
plt.title('Survival by Passenger Class')
plt.show()
```

## Plot the graph with respect to survival counts

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived',hue='Survived', data=data, palette='viridis')
plt.title('Survival Counts')
plt.show()
```

## Age distribution of survivors vs non-survivors


```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data[data['Survived'] == 1]['Age'], label='Survived', fill=True, color='green')
sns.kdeplot(data=data[data['Survived'] == 0]['Age'], label='Did Not Survive', fill=True, color='red')
plt.title('Age Distribution: Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```
